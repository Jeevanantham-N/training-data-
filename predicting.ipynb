```
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/7_logistic_reg/Exercise/HR_comma_sep.csv")
df
```

```
dummy_dept = pd.get_dummies(df.Department)
dummy_salary = pd.get_dummies(df.salary)
dummy_salary
```

```
merged = pd.concat([df, dummy_dept], axis = "columns")
merged
```

```
merged1 = pd.concat([merged, dummy_salary], axis = "columns")
merged1
```

```
final = merged1.drop(["Department", "salary", "hr", "high"], axis = "columns")
final
```

```
y = final.left
X = final.drop(columns=["left"], axis = "columns")
X
```

```
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style of seaborn for better visualization
sns.set(style="whitegrid")

# Create a bar chart
plt.figure(figsize=(8, 5))
sns.countplot(x="salary", hue="left", data=df)

# Add labels and title
plt.xlabel("Salary")
plt.ylabel("Number of Employees")
plt.title("Impact of Employee Salaries on Retention")
plt.legend(title="Left", labels=["Not Left", "Left"])

# Show the plot
plt.show()
```

```
sns.set(style = "whitegrid")
plt.figure(figsize=(12, 5))

sns.countplot(hue="left", x="Department", data=df)

plt.xlabel("Departments")
plt.ylabel("Left")
plt.legend(title = "Left", labels = ["Not Left", "Left"])
plt.title("Correlation of Departments and Employee Retention")

plt.show()
```

```
from sklearn.model_selection import train_test_split
from sklearn import linear_model

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
model = linear_model.LogisticRegression()

model.fit(X_train, y_train)
```

```
model.predict(X_test)
```

```
model.score(X_test, y_test)
```
